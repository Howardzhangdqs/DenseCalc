```python
def size_to_embedding(size: int) -> int:
    if size == 16:
        return 0
    elif size == 32:
        return 1
    elif size == 64:
        return 2
    elif size == 128:
        return 3
    elif size == 256:
        return 4
    elif size == 512:
        return 5
    elif size == 1024:
        return 6
```
